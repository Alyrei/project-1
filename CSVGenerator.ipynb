```python
import pandas as pd
import requests
from nba_api.stats.static import teams
from nba_api.stats.library import data
from nba_api.stats.endpoints import leaguegamefinder

#Creating list of Team Abbreviations
nba_teams = teams.get_teams()
nba_df = pd.DataFrame(nba_teams)
team_ids = nba_df["id"]

nba_df
```

```python
# This code works but only pulls ONE team at a time, gotta write a script to pull all 30 at once

ATL = leaguegamefinder.LeagueGameFinder(team_id_nullable=team_ids[0])
ATL_games = ATL.get_data_frames()[0]
ATL_games['SEASON_ID'] = ATL_games['SEASON_ID'].map(lambda x: str(x)[1:])
ATL_seasons = ATL_games.set_index("SEASON_ID")
ATL_ten_years = ATL_seasons[ATL_seasons.index.isin(['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018'])]
ATL_ten_years.to_csv("./data/ATL.csv")
```
