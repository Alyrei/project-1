```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import glob
import numpy as np
import os 

path = "./Data"
all_files = glob.glob(os.path.join(path, "*.csv"))

df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)

path_two = "./Playoffs"
all_files_two = glob.glob(os.path.join(path_two, "*.csv"))

df_from_each_file_two = (pd.read_csv(f) for f in all_files_two)
concatenated_two_df   = pd.concat(df_from_each_file_two, ignore_index=False)
concatenated_two = concatenated_two_df.sort_values(by=["Year","Game"])

playoff_csv = "PlayOffs.csv"
playoff_df = pd.read_csv(playoff_csv)
playoff_df = playoff_df.dropna(1,"all")

one labels =['SEASON_ID' 'TEAM_ID' 'TEAM_ABBREVIATION' 'TEAM_NAME' 'GAME_ID'
 'GAME_DATE' 'MATCHUP' 'WL' 'MIN' 'PTS' 'FGM' 'FGA' 'FG_PCT' 'FG3M' 'FG3A'
 'FG3_PCT' 'FTM' 'FTA' 'FT_PCT' 'OREB' 'DREB' 'REB' 'AST' 'STL' 'BLK'
 'TOV' 'PF' 'PLUS_MINUS']
two labels =['Year' 'Team' 'Game' 'Win' 'Home' 'MP' 'FG' 'FGA' 'FGP' 'TP' 'TPA' 'TPP'
 'FT' 'FTA' 'FTP' 'ORB' 'DRB' 'TRB' 'AST' 'STL' 'BLK' 'TOV' 'PF' 'PTS']
```

```python
season_data = concatenated_df
season_data = season_data.set_index(["TEAM_NAME"]+["SEASON_ID"])
season_data.sort_values(by=["TEAM_NAME","SEASON_ID"])
season_group = season_data.groupby(["TEAM_NAME","SEASON_ID"]).agg({"PTS":"mean","FGM":"mean","FGA":"mean",
                                                                   "FG_PCT":"mean","FG3M":"mean","FG3A":"mean",
                                                                   "FG3_PCT":"mean","FTM":"mean","FTA":"mean",
                                                                   "FT_PCT":"mean","OREB":"mean","DREB":"mean",
                                                                   "REB":"mean","AST":"mean","STL":"mean",
                                                                   "BLK":"mean","TOV":"mean","PF":"mean",
                                                                   "PLUS_MINUS":"count"})
season_group_df = season_group.style.format({
    'PTS':'{:,.2f}'.format,'FGM':'{:,.2f}'.format,'FGA':'{:,.2f}'.format, 
    'FG_PCT':'{:,.2%}'.format,'FG3M':'{:,.2f}'.format,'FG3A':'{:,.2f}'.format,
    'FG3_PCT':'{:,.2%}'.format,'FTM':'{:,.2f}'.format,'FTA':'{:,.2f}'.format, 
    'FT_PCT':'{:,.2%}'.format,'OREB':'{:,.2f}'.format,'DREB':'{:,.2f}'.format,
    'REB':'{:,.2f}'.format,'AST':'{:,.2f}'.format,'STL':'{:,.2f}'.format,
    'BLK':'{:,.2f}'.format,'TOV':'{:,.2f}'.format,
})
# len(season_data["WL"].loc[season_data["WL"] == "W", :])
season_group_df
```
